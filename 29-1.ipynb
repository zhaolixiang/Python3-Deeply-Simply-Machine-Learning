```python
# 导入pandas
import pandas as pd

# 手工输入一个数据表
fruits = pd.DataFrame({'数值特征': [5, 6, 7, 8, 9], '类型特征': ['西瓜', '香蕉', '橘子', '苹果', '葡萄']})
# 显示fruits 数据表
display(fruits)


# 转化数据表中的字符串为数值
fruits_dum=pd.get_dummies(fruits)
# 显示转化后的数据表
display(fruits_dum)


# 令程序将数值也看作字符串
fruits['数值特征']=fruits['数值特征'].astype(str)
# 在用get_dummies转化字符串
pd.get_dummies(fruits,columns=['数值特征'])

```
